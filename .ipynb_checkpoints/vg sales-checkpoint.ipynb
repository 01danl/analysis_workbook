```python
import pandas as pd

data = pd.read_csv('csv files/vgsales.csv')
```

```python
data.head()
```

### Find the total sales in North America, Europe, Japan and other regions for all games.

```python
data['Total'] = data['NA_Sales'] + data['EU_Sales'] + data['JP_Sales'] + data['Other_Sales'] 
data['Total']
```

```python
data.pop('Total')
```

### Calculate the Difference Between Regional and Global Sales

```python
sum_regional_sales = data['NA_Sales'] + data['EU_Sales'] + data['JP_Sales'] + data['Other_Sales'] 
difference = sum_regional_sales - data['Global_Sales']
print(difference)
```

### Find the Top 5 Best-Selling Games Globally

```python
data.sort_values(by='Global_Sales', ascending = False).head(5)
```

### Find the Total Sales for Each Genre

```python
data['Genre'].unique() #to know how much genres we have
```

```python
genre_sales = data.groupby('Genre')['Global_Sales'].sum() 
genre_sales
```
